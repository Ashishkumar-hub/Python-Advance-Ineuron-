Q1. Explain three-dimensional data indexing.

Answer: In Python, portions of data can be accessed using indices, slices, column headings, and condition-based subsetting. Python uses 0-based indexing, in which the first element in a list, tuple or any other data structure has an index of 0. A simple example is shown below

```python
d3_array = np.array([[[0, 1],
                      [2, 3]],

                      [[4, 5],
                      [6, 7]]])
```

```python
'''Layout: 

sheet 0:
       [0, 1]
       [2, 3]

sheet 1:
       [4, 5]
       [6, 7] '''
```

```python
d3_array[0, 1, 0] #Thus d3_array[0, 1, 0] specifies the element residing in sheet-0, at row-1 and column-0
```

```python
d3_array[:, 0, 0] #d3_array[:, 0, 0] specifies the elements in row-0 and column-0 of both sheets:
```

```python
d3_array[1] #d3_array[1], which recall is shorthand for d3_array[1, :, :], selects both rows and both columns of sheet-1
```


Q2. What's the difference between a series and a dataframe?

Answer: Series is a type of list in pandas which can take integer values, string values, double values and more. But in Pandas Series return an object in the form of list, having index starting from 0 to n, Where n is the length of values in series.

Series can only contain single list with index, whereas dataframe can be made of more than one series or we can say that a dataframe is a collection of series that can be used to analyse the data. Example: 

```python
import pandas as pd
import matplotlib.pyplot as plt
author = ['Jitender', 'Purnima', 'Arpit', 'Jyoti']
article = [210, 211, 114, 178]
auth_series = pd.Series(author)
article_series = pd.Series(article)
frame = { 'Author': auth_series, 'Article': article_series }
result = pd.DataFrame(frame)
age = [21, 21, 24, 23]
result['Age'] = pd.Series(age)
print(result)
```

Q4. How do you use pandas to make a data frame out of n-dimensional arrays?

Answer: To convert an array to a dataframe with Python we need to: 
1. Have NumPy array (e.g., np_array) 
2. Use the pd.DataFrame() constructor like this: df = pd.DataFrame(np_array, columns=[‘Column1’, ‘Column2’]). 

Note: Each column in NumPy array needs to be named with columns. 

```python
import numpy as np
import pandas as pd
numpy_array = np.array([[1,"ashish"],[2,"ashu"],
                        [3,"vikash"],[4,"rohit"]])
df = pd.DataFrame(numpy_array, index=['day1', 'day2', 'day3', 'day4'],
                 columns=['digits', 'words'])
df
```

Q5. Explain the notion of pandas plotting.

Answer: Pandas uses the plot() method to create diagrams. We can use Pyplot, a submodule of the Matplotlib library to visualize the diagram on the screen.

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2022", periods=1000))
ts = ts.cumsum()
ts.plot()
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list("ABCD"))
df = df.cumsum()
df.plot()
```
